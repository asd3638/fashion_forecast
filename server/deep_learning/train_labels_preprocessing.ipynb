```python
# 전체 흐름 참고: https://www.analyticsvidhya.com/blog/2019/04/build-first-multi-label-image-classification-model-python/
from ast import literal_eval

import glob2
import pandas as pd
import json
import csv
import glob2
%matplotlib inline
```

```python
# "라벨링데이터" 하위에 있는 모든 json파일 읽어오기
# https://blr.design/blog/python-multiple-json-to-csv/
files = glob2.glob('C:\k-fashion-dataset/Training/labels/*/*', recursive=True)
less_files = files[:30000]
print(len(less_files))
```

```python
data = []
for single_file in less_files:
    with open(single_file, 'r', encoding='utf-8') as f:
        # Use 'try-except' to skip files that may be missing data
        try:
            json_file = json.load(f)
            file_id = json_file["데이터셋 정보"]["파일 번호"]
            file_name = json_file["데이터셋 정보"]["파일 이름"]
            top = json_file["데이터셋 정보"]["데이터셋 상세설명"]["라벨링"]["상의"][0]
            data.append([file_id, file_name, top])
            print("saved", single_file[70:])
        except KeyError:
            print(f'Skipping {single_file}')
# Add header
data.insert(0, ['ID', 'File Name', 'Top'])
print("ALL DONE!")
print(len(data))
```

```python
# csv로 변환
csv_filename = 'train.csv'
with open(csv_filename, "w", encoding='utf-8', newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

print("Updated CSV")
```

```python
# csv칼럼 내부 json 객체 flatten하기
# https://stackoverflow.com/questions/39899005/how-to-flatten-a-pandas-dataframe-with-some-columns-as-json
train = pd.read_csv('train.csv', converters = {'Top': literal_eval}, encoding='utf-8') # reading the csv file
```

```python
train
```

```python
TOP = pd.json_normalize(train["Top"]).add_prefix('Top.')
train = train[['ID', 'File Name']].join([TOP])
train
```

```python
# flatten해서 생긴 칼럼들 중 우리 모델에 필요 없는 칼럼들 없애기
pd.set_option('display.max_columns', None)
train = train.drop(columns=['Top.서브색상', 'Top.색상',
                            'Top.소재', 'Top.프린트', 'Top.핏',
                            'Top.디테일', 'Top.기장',
                            'Top.넥라인', 'Top.옷깃'
                            ])
train
```

```python
# one-hot 인코딩 하기
columns = train.columns[2:]
train = pd.get_dummies(train, columns=columns, prefix_sep=".")
```
